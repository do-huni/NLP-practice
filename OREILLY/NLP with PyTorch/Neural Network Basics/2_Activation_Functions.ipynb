```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sigmoid function
x = torch.range(-5.,5,0.1)
y = torch.sigmoid(x)
plt.plot(x.numpy(), y.numpy())
plt.show()
```

```python
#hyperbolic tangent(tanh) function
x = torch.range(-5.,5,0.1)
y = torch.tanh(x)
plt.plot(x.numpy(), y.numpy())
plt.show()
```

```python
#Rectified Linear Unit(ReLU) function
x = torch.range(-5.,5,0.1)
y = torch.relu(x)
plt.plot(x.numpy(), y.numpy())
plt.show()
```

```python
#softmax function
softmax =torch.nn.Softmax(dim=1)
x = torch.randn(1,100)
y = softmax(x)
plt.plot(x.numpy(), y.numpy(), 'o')
plt.show()
print(f'sum: {torch.sum(y)}"')
```
